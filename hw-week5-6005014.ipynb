```python
from astroquery.sdss import SDSS
```

```python
query = "SELECT top 100 p.petroMag_g,s.z, p.ObjID,s.zErr,p.petroMagErr_g       FROM PhotoObj p         JOIN specObj s ON s.bestObjID=p.objID       WHERE s.zWarning = 0 and s.z between 0.3 and 0.4"
res=SDSS.query_sql(query)
print(res[:20])
```

```python
fig = plt.figure(figsize=(50,15))
ax1= fig.add_subplot(121)
ax1.errorbar(x=res['z'] , y=res['petroMag_g'] ,xerr=res['zErr'], yerr=res['petroMagErr_g'] , color='black',ecolor='g',fmt='o' )
ax1.set_xlabel('z')
ax1.set_ylabel('petroMag_g')
ax1.set_title('photoObj and specObj x,y error')
```

```python
fig = plt.figure(figsize=(28,15))
ax1= fig.add_subplot(121)
ax1.scatter(x=res['z'] , y=res['petroMag_g'] , color='black' )
ax1.set_xlabel('z')
ax1.set_ylabel('petroMag_g')
ax1.set_title('photoObj and specObj no error')
```
